```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.random.normal(0,1,250000)
```

```python
plt.hist(a,bins=10000);
plt.show()
```

```python
a1=np.random.default_rng(12).integers(low=5,high=500,size=25)
a2=np.random.default_rng(102).integers(low=10,high=50,size=25)
plt.title("Random Plot")
plt.xlabel("A1")
plt.ylabel("A2")
plt.bar(a1,a2,width=5)
```

```python
a1=np.random.default_rng(12).integers(low=5,high=500,size=25)
a2=np.random.default_rng(102).integers(low=10,high=50,size=25)
plt.title("Random Plot")
plt.xlabel("A1")
plt.ylabel("A2")
plt.barh(a1,a2,height=5,color="black")
```
